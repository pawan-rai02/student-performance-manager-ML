```python
# import data and libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")

```

```python
df = pd.read_csv(r'data\StudentsPerformance.csv')
```

```python
df.head()
```

```python
# dataset info
'''
- gender : male or female
- race/ethnicity : group A, B, C, D, E
- parental level of education : some high school, high school, some college, associate's degree, bachelor's degree, master's degree

- lunch : standard or free/reduced
- test preparation course : none or completed before test
- math score : 0-100
- reading score : 0-100
- writing score : 0-100
'''
```

```python
# exploring data
print('categories in gender:', df['gender'].nunique())
print('categories in race/ethnicity:', df['race/ethnicity'].nunique())
print('categories in parental level of education:', df['parental level of education'].nunique())
print('categories in lunch:', df['lunch'].nunique())
print('categories in test preparation course:', df['test preparation course'].nunique())
```

```python
# define numerical and categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print cols
print('numerical features:', numeric_features)
print('categorical features:', categorical_features)
```

```python
# adding cols for total score and average score
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
df['average'] = df['total score'] / 3

df.head()
```

```python
# full marks 
reading_full = df[df['reading score'] == 100].count()
writing_full = df[df['writing score'] == 100].count()
math_full = df[df['math score'] == 100].count()

print('number of students with full marks in reading:', reading_full['reading score'])
print('number of students with full marks in writing:', writing_full['writing score'])
print('number of students with full marks in math:', math_full['math score'])
```

```python
reading_less_20 = df[df['reading score'] < 20].count()
writing_less_20 = df[df['writing score'] < 20].count()
math_less_20 = df[df['math score'] < 20].count()

print('number of students with less than 20 marks in reading:', reading_less_20['reading score'])
print('number of students with less than 20 marks in writing:', writing_less_20['writing score'])
print('number of students with less than 20 marks in math:', math_less_20['math score'])
```

```python
# histogram and kde
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

sns.histplot(df['average'], kde=True, ax=axs[0])
axs[0].set_title('Distribution of Average Scores')

sns.histplot(data = df, x = 'average', hue= 'gender', kde=True, ax=axs[1])
axs[1].set_title('Distribution of Average Scores by Gender')

```

```python
# histogram and kde
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

sns.histplot(df['total score'], kde=True, ax=axs[0])
axs[0].set_title('Distribution of Total Scores')

sns.histplot(data = df, x = 'total score', hue= 'gender', kde=True, ax=axs[1])
axs[1].set_title('Distribution of Total Scores by Gender')
```
